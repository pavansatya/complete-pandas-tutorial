```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/Olympics-Dataset/master/athletes/bios.csv')
df.shape
```

```python
df.head()
```

1) Cleaning birth date and death date columns

```python
df['birth_date'] = df.Born.str.split(' in').str[0]
```

```python
df['birth_date'] = pd.to_datetime(df['Born'], format='%d %B %Y', errors='coerce')
```

```python
# df.drop(columns=['Born'], inplace=True)
```

```python
df['death_date'] = df.Died.str.split(' in').str[0]
```

```python
df['death_date'] = pd.to_datetime(df['Died'], format='%d %B %Y', errors='coerce')
```

```python
# df.drop(columns=['Died'], inplace=True)
```

2) Cleaning Height and Weights from Measurements column

```python
df['Height'] = df.Measurements.str.split(' cm /').str[0]
```

```python
df.Height = pd.to_numeric(df.Height, errors='coerce').astype('Int64')
```

```python
df.Height.isna().sum()
```

```python
df['Weight'] = df.Measurements.str.split(' cm /').str[1].str.split(' kg').str[0]
df.Weight = pd.to_numeric(df.Weight, errors='coerce').astype('Int64')
```

```python
df.Weight.isna().sum()
```

```python
# df.drop(columns=['Measurements'], inplace=True)
```

```python
df.tail()
```

3) get rid of bullet points in columns `Full name` and `Used name`

```python
df['Full name'] = df['Full name'].str.replace('•', ' ')
```

```python
df['Used name'] = df['Used name'].str.replace('•', ' ')
```

```python
# This is other way to do it 
# df[['height', 'weight']] = df.Measurements.str.split(' /', expand=True).head()
```

```python
city_pattern = r'in ([\w]+)' 
region_pattern = r', ([\w ]+)' 
country_pattern = r'\((\w{3})\)$'

df['city'] = df.Born.str.extract(city_pattern, expand=False)
df['region'] = df.Born.str.extract(region_pattern, expand=False)
df['country'] = df.Born.str.extract(country_pattern, expand=False) 
```
